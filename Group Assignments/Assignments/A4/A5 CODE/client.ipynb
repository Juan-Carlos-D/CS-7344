```python
import socket

# Same XOR key used by the server
KEY = 23  

def encrypt(message: bytes) -> bytes:
    return bytes([b ^ KEY for b in message])

def decrypt(message: bytes) -> bytes:
    return bytes([b ^ KEY for b in message])

# Create TCP socket
client = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
client.connect(('127.0.0.1', 9000))

print("Connected to server. Type messages below. Type 'quit' to exit.")

while True:
    msg = input("Enter message: ")

    if msg.lower() == "quit":
        break

    encrypted = encrypt(msg.encode())
    client.send(encrypted)

    response = client.recv(1024)
    decrypted_response = decrypt(response).decode()
    print("Server:", decrypted_response)

client.close()
```
