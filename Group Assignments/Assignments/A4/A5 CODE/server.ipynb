```python
import socket

# Simple XOR key for symmetric encryption/decryption
KEY = 23  

def encrypt(message: bytes) -> bytes:
    return bytes([b ^ KEY for b in message])

def decrypt(message: bytes) -> bytes:
    return bytes([b ^ KEY for b in message])

# Create TCP socket
server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind(('0.0.0.0', 9000))
server.listen(1)

print("Server running on port 9000... Waiting for connection.")

conn, addr = server.accept()
print("Connected to:", addr)

while True:
    data = conn.recv(1024)
    if not data:
        break  # client disconnected

    decrypted_msg = decrypt(data).decode()
    print("Client said:", decrypted_msg)

    response_text = "Received: " + decrypted_msg
    encrypted_response = encrypt(response_text.encode())
    conn.send(encrypted_response)

conn.close()
server.close()
```
